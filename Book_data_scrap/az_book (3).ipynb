### loop through all pages of the website 

```python
book_list = []
for page in range(1, 14):
    url = f'https://www.az-bookstore.com/page/{page}/?post_type=product'
    result = requests.get(url, headers={"User-Agent":"Chrome/103.0.0.0"})
    soup = bts(result.text, "html.parser")
    ul = soup.find('ul', {'class':'products columns-6'})
    books = ul.find_all('li')
    for book in books:
        title = book.find("h3").find('a').text.strip()
        author = book.find("div", class_ = 'product-author').text.replace('by','').strip() 
        if page == 4 and title == 'Harry Potter The Complete Collection':
            price = 59
            book_list.append([title, author, price])
            continue
        price = book.find('span', class_ ='price')
        price = float(price.text[1:]) if price else None
        book_list.append([title, author, price])
```

```python
# loop through all pages of the website
book_list = []
for page in range(1, 3):
    url = f'https://www.az-bookstore.com/page/{page}/?post_type=product'
    result = requests.get(url, headers={"User-Agent":"Chrome/103.0.0.0"})
    soup = bts(result.text, "html.parser")
    ul = soup.find('ul', {'class':'products columns-6'})
    books = ul.find_all('li')
    for book in books:
        title = book.find("h3").find('a').text.strip()
        author = book.find("div", class_ = 'product-author').text.replace('by','').strip() 
        if page == 4 and title == 'Harry Potter The Complete Collection':
            price = 59
            book_list.append([title, author, price])
            continue
        price = book.find('span', class_ ='price')
        price = float(price.text[1:]) if price else None
        book_list.append([title, author, price])
        
    title_list = [book.find("h3").find('a').text.strip().lower().replace(' ', '-') for book in books]
    
    for index, i in enumerate(title_list):
        url1 = f'https://www.az-bookstore.com/product/{i}/'
        result1 = requests.get(url1, headers={"User-Agent":"Chrome/103.0.0.0"})
        soup1 = bts(result1.text, "html.parser")
        article = soup1.find('article', class_ = 'post_item post_item_single post_item_product')
        category = soup1.find('div', class_ = 'product_meta').find('span', class_ = 'posted_in').find('a').text.strip() if category else None
        book_condition = soup1.select_one('div.woocommerce-product-details__short-description p br')
        book_list[index].append([category, book_condition])
```
