# This is the final code that we use to scrap data from Viator

Be aware that It took a really long time to run this code, so to run this code, reducing the range is recommended

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

lst = []
#101
for i in range (2, 101):
    url = f"https://www.viator.com/Cambodia/d12-ttd/{i}"
    response = requests.get(url)
    response = response.content

    soup = BeautifulSoup(response, 'html.parser')
    ol = soup.find('div', class_='product-list product-list-rows mt-3 mt-md-0')
    
    elements = ol.find_all('div', class_='tracked-elements')
    for element in elements:
        
        details_url = element.find('a')['href']
        if not details_url.startswith('http'):
            details_url = 'https://www.viator.com' + details_url
            
        details_r = requests.get(details_url)
        details_s = BeautifulSoup(details_r.content, 'html.parser')
        
# Data that we scrap:

        tour_n = getattr(element.find('a', class_='text-dark highlight-able card-link'), 'text', 'N/A')
        tour_type = details_s.select_one('div.itemWrapper__3IOj.childItem__3EYW a.navigationLink__nipx.bold__324P')
        if tour_type is not None:
            tour_type_text = str(tour_type.text.strip())
        else:
            tour_type_text = 'N/A'
        num_reviewer = getattr(element.find('span', class_='small text-body text-nowrap reviewLink product-list-card-rating-label'), 'text', 'N/A')
        rating = getattr(element.find('svg', class_='d-flex align-self-center mr-2 fill-yellow'), 'text', 'N/A')
        duration = getattr(details_s.find('div', class_='item__3eVq'), 'text', 'N/A')
        price = getattr(element.find('div', class_='h3 line-height-same mb-0 price-font text-md-right mr-md-0'), 'text', 'N/A')
        province = getattr(details_s.find('span', class_='location__2Gmy'), 'text', 'N/A')
        lis = details_s.select_one('ul.featureList__34Y_.noBullet__2pPd')
        
#Display data:

        try:
            print(f"Tour Name: {tour_n.text}")
        except AttributeError:
            print(f"Tour Name: {tour_n}")
        
        print(f"Tour Type: {tour_type_text}")
            
        try:
            print(f"Number of Revewer: {num_reviewer.text}")
        except AttributeError:
            print(f"Number of Revewer: {num_reviewer}")
        
        try:
            print(f"Rating: {rating.text}")
        except AttributeError:
            print(f"Rating: {rating}")
       
        try:
            print(f"Duration: {duration.text}")
        except AttributeError:
            print(f"Duration: {duration}")
        
        try:
            print(f"Price: {price.text}")
        except AttributeError:
            print(f"Price: {price}")
            
    
        try:
            print(f"Location: {province.text}")
        except AttributeError:
            print(f"Location: {province}")
            
        print(f"Hightlight: ", end='')
        if lis is not None:
            highlight_list = [li.text.strip() for li in lis]
            highlight_string = ', '.join(highlight_list)
            print(highlight_string)
            #print("\n")
        else:
            print("N/A")
            print("\n")
            
        print("\n")
        
        
# Append each variable into a list 
        
        lst.append([tour_n, tour_type, num_reviewer, rating, duration, price, highlight_string, province])
        
```

<h4>Convert the data that we scrapped into csv format</h4>

```python
import pandas as pd
df = pd.DataFrame(lst, columns=['Tour_Name','Tour_Type', 'Number_of_Reviewer', 'Rating' , 'Duration', 'Price', 'Highlight', 'Location'])
#df.to_csv("C:/Users/Ds Computer/Data/tour_data_tripadviser.csv", encoding='utf-8')
df.to_csv("tour_data_viator.csv")
df
```

# This is the final code that we use to scrap data from Trip Adviser

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import html

lst = []
#101
for i in range (2, 101):
    url = f"https://www.viator.com/Cambodia/d12-ttd/{i}"
    response = requests.get(url)
    response = response.content

    soup = BeautifulSoup(response, 'html.parser')
    ol = soup.find('div', class_='product-list product-list-rows mt-3 mt-md-0')
    
    elements = ol.find_all('div', class_='tracked-elements')
    for element in elements:
        
        details_url = element.find('a')['href']
        if not details_url.startswith('http'):
            details_url = 'https://www.viator.com' + details_url
            
        details_r = requests.get(details_url)
        details_s = BeautifulSoup(details_r.content, 'html.parser')
        
        # Data that we scrap:
        tour_n = getattr(element.find('a', class_='text-dark highlight-able card-link'), 'text', 'N/A')
        tour_type = details_s.select_one('div.itemWrapper__3IOj.childItem__3EYW a.navigationLink__nipx.bold__324P')
        if tour_type is not None:
            tour_type_text = html.unescape(tour_type.text.strip())
        else:
            tour_type_text = 'N/A'
        num_reviewer = getattr(element.find('span', class_='small text-body text-nowrap reviewLink product-list-card-rating-label'), 'text', 'N/A')
        rating = getattr(element.find('svg', class_='d-flex align-self-center mr-2 fill-yellow'), 'text', 'N/A')
        duration = getattr(details_s.find('div', class_='item__3eVq'), 'text', 'N/A')
        price = getattr(element.find('div', class_='h3 line-height-same mb-0 price-font text-md-right mr-md-0'), 'text', 'N/A')
        province = getattr(details_s.find('span', class_='location__2Gmy'), 'text', 'N/A')
        lis = details_s.select_one('ul.featureList__34Y_.noBullet__2pPd')
        
        # Create a row containing all the scraped values
        row = [tour_n, tour_type_text, num_reviewer, rating, duration, price, 'N/A', province]
        if lis is not None:
            highlight_list = [li.text.strip() for li in lis]
            highlight_string = ', '.join(highlight_list)
            row[6] = highlight_string
        else:
            print("N/A")
            

        # Append the row to the lst list only if it is not already in the list
        if row not in lst:
            lst.append(row)
        
```

```python
df = pd.DataFrame(lst, columns=['Tour_Name','Tour_Type', 'Number_of_Reviewer', 'Rating' , 'Duration', 'Price', 'Highlight', 'Location'])
df.to_csv("tour_data_viator.csv")
df
```
