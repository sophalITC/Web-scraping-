```python
count = 1
data = 0
for month in checkMonth:
    for day in checkday:
        t=46
        url = f"https://www.booking.com/hotel/kh/marvel-holiday-boutique-villa.ms.html?aid=356980&label=gog235jc-1FCAsoeEIdbWFydmVsLWhvbGlkYXktYm91dGlxdWUtdmlsbGFIM1gDaHiIAQGYARu4ARfIAQzYAQHoAQH4AQyIAgGoAgO4AsDr2aIGwAIB0gIkOWMzZDk3ZjEtNDU4Yi00MjAwLWEwYWQtYTJmMDAxYTZiNzkw2AIG4AIB&sid=42cbc141abd816955d8fbe7ce58d0863&checkin_month={month}&checkin_monthday={day}&checkin_year=2023&checkout_month={month}&checkout_monthday={day+1}&checkout_year=2023&dist=0&do_availability_check=1&group_adults=2&group_children=0&hp_avform=1&hp_group_set=0&no_rooms=1&origin=hp&sb_price_type=total&src=hotel&stay_on_hp=1&type=total&#availability_target"

        headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"}
        target_url = url
        resp = requests.get(target_url, headers=headers)
        #print(resp.status_code)

        g=list()
        k={}
        soup = BeautifulSoup(resp.text, 'html.parser')

        name=soup.find("h2",{"class":"pp-header__title"}).text
        address=soup.find("span",{"class":"hp_address_subtitle"}).text.strip("\n")
        rating=soup.find("div",{"class":"d10a6220b4"}).text
            
        ids= []
        try:
            tr = soup.find_all("tr")
        except:
            tr = None

        for y in range(0,len(tr)):
            try:
                id = tr[y].get('data-block-id')

            except:
                id = None

            if( id is not None):
                ids.append(id)
        #print(ids)
        print("ids are ",len(ids))
        data = data + len(ids)

        for i in range(0,len(ids)):
            try:
                k['Name'] = name
                k['Adress'] = address
                k['Rating'] = rating
                k['URL'] = target_url
                allData = soup.find("tr",{"data-block-id":ids[i]})
                try:
                    rooms = allData.find("span",{"class":"hprt-roomtype-icon-link"})
                except:
                    rooms=None
                if(rooms is not None):
                    last_room = rooms.text.replace("\n","")
                try:
                    k["RoomType"]=rooms.text.replace("\n","")
                except:
                    k["RoomType"]=last_room
                #print(k["RoomType"])
                
                try:
                    roomSize = allData.find("div", {"data-name-en":"room size"})
                except:
                    roomSize=None
                if(rooms is not None):
                    last_roomSize = roomSize.text.replace("\n","")
                try:
                    k["RoomSize"]=roomSize.text.replace("\n","")
                except:
                    k["RoomSize"]=last_roomSize
                #print(k["RoomSize"])
                
                price = allData.find("div",{"class":"bui-price-display__value prco-text-nowrap-helper prco-inline-block-maker-helper prco-f-font-heading"})
                k["Price"]=price.text.replace("\n","")
                #print(k["Price"])
                
                
                
                g.append(k)
                k={}
            except:
                k['Name'] = None
                k['Adress'] = None
                k['Rating'] = None
                k["RoomType"]=None
                k["Price"]=None
                k["RoomSize"] = None
                k["URL"] = None
                
        # Add to dataframe
        df = pd.DataFrame(g)
        
        # Make csv file
        fileName = df.to_csv(f'hotel{t}_{day}_{month}.csv')
        print(count)
        print(f'Done: {day}/{month}')
        count = count + 1

print(f'Number of row: {data}')
        
```
