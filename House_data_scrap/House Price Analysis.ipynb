```python
# Import Libraries
from bs4 import BeautifulSoup
from urllib import request
from urllib.request import Request, urlopen
from csv import writer
```

```python
#1st page

# Define the website URL and send a request to the website with a user-agent header
url = "https://www.realestate.com.kh/buy/phnom-penh/house/?active_tab=popularLocations&categories=House&order_by=relevance&property_type=residential&q=location%3A%20Phnom%20Penh&search_type=sale"
request_site = Request(url, headers ={'User-Agent': 'Mozilla/5.0'} )
# read in the HTML content of the website
webpage = urlopen(request_site).read()
# parse the HTML content using BeautifulSoup
soup = BeautifulSoup(webpage, 'html.parser')  

# Open a CSV file for writing and write the header row
with open('House Price Analysis.csv','a',encoding='utf8', newline='') as f:
    thewriter = writer(f) 
    header = ['Type','Price', 'Location', 'Bedrooms', 'Bathrooms'] 
    thewriter.writerow(header)

# Open the same CSV file for writing again and append the data rows
with open('House Price Analysis.csv','a',encoding='utf8', newline='') as f:
    thewriter = writer(f)
    
     # Loop through each listing on the webpage
    for i in soup.find_all('div', class_='copy-wrapper'):
         # Extract the type of house from the heading element
        heading = i.find('div', class_='heading')
        if heading is not None:
            whatType = heading.text.strip()
            typeOfHouse = whatType.split('$')[0] 
             # Extract the price from the price element
            for j in i.find_all('div', class_='heading'):
                for k in j.find_all('div', class_='price'):
                    perPrice = k.find('span', class_='value').text.strip()
                    
             # Extract the location from the address element    
            for l in i.find_all('div', class_='address-features-block'):
                tmpLoc = l.find('div', class_='address').text.strip()
                perLoc = tmpLoc.split('')[1]
                
             # Extract the number of bedrooms from the bedroom element
            for br in i.find_all('div', class_='private'):
                tmpBr = br.find('div', class_='feature').text.strip()
                perBr = tmpBr[len(tmpBr)-1]
                
             # Extract the number of bathrooms from the bathroom element
            for ba in i.find_all('div', class_='private'):
                tmpBa = ba.find('div', class_='feature').text.strip()
                perBat = tmpBa[len(tmpBa)-1]
            
             # Write the extracted information as a new row in the CSV file
            info = [typeOfHouse, perPrice, perLoc, perBr, perBat]
            thewriter.writerow(info)
            typeOfHouse = ""
            perPrice  = ""
    
```
