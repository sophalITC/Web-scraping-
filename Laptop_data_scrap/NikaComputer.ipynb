```python
# Asus laptop
from bs4 import BeautifulSoup
import requests
import pandas as pd


url = "https://nika2u.com/public/shop/nika_computer?sub_id=subcat612af0581aed73e114351e4326749156"
response = requests.get(url)


soup = BeautifulSoup(response.content, 'lxml')


products = soup.find_all('div', class_='product-item')

price_=[]
model_=[]
CPU_=[]
RAM_=[]
Storage_=[]
Display_=[]
Graphic_=[]
OS_=[]
Speaker_=[]
Keyboard_=[]
Battery_=[]
Weight_=[]

for product in products:
    model = product.find('h4', class_='title').text.strip()
    price = product.find('span', class_='price').text.strip()
    detail = product.find('p', class_='description')
    
    model_.append(model)
    price_.append(price)
    CPU=detail.text.split('▪️')[1].strip()
    CPU_.append(CPU)

    RAM=detail.text.split('▪️')[2].strip()
    RAM_.append(RAM)

    Storage=detail.text.split('▪️')[3].strip()
    Storage_.append(Storage)

    Display=detail.text.split('▪️')[4].strip()
    Display_.append(Display)

    Graphic=detail.text.split('▪️')[5].strip()
    Graphic_.append(Graphic)

    OS=detail.text.split('▪️')[6].strip()
    OS_.append(OS)

    Speaker=detail.text.split('▪️')[7].strip()
    Speaker_.append(Speaker)

    Keyboard=detail.text.split('▪️')[8].strip()
    Keyboard_.append(Keyboard)
    
    Weight=detail.text.split('▪️')[9].strip()
    Weight_.append(Weight)
dict={'Model':model_,'Price':price_,'CPU':CPU_,'RAM':RAM_,'Storage':Storage_,'Display':Display_,'Graphic':Graphic_,'OS':OS_,'Speaker':Speaker_,'Keyboard':Keyboard_,'Weight':Weight_}
df3=pd.DataFrame(dict)
df3.to_csv("1111.csv")
```
