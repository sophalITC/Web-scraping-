# First of all, let's just try to scrape one page of the IMDB website

### Import the required libraries

````python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
from time import sleep
from random import randint
````

````python
url = "https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating"
response = requests.get(url) # get the response from the server of the web page
soup = BeautifulSoup(response.content, 'html.parser') 
````

# Scraping the name, time, meta score, year, rating

# Declaring the empty lists

````python
movie_name = []
year = []
rating = []
time = []
metascore = []
votes = []
gross = []
````

````python
movie_data = soup.find_all('div',attrs={'class':"lister-item mode-advanced"})
movie_data
````

````python
len(movie_data)
````

````python
for store in movie_data:
    name = store.h3.a.text # store the name of the movie
    movie_name.append(name)
    

    year_of_release = store.h3.find('span', class_="lister-item-year text-muted unbold").text.replace('(','').replace(')','')
    year.append(year_of_release)


    runtime = store.p.find('span',class_='runtime').text.replace(' min', '') if store.p.find('span',class_='runtime') else "****"
    time.append(runtime)

    rate = store.find('div', class_="inline-block ratings-imdb-rating").text.replace('\n', '')
    rating.append(rate)


    meta = store.find('span', class_='metascore').text.rstrip() if store.find('span', class_='metascore') else '^^^^^^^^'
    metascore.append(meta)


    value = store.find_all('span', attrs = {'name': 'nv'})
    vote = value[0].text if len(value) > 1 else '*****' 
    grosses = value[1].text if len(value) > 1 else '*****'
    votes.append(vote)
    gross.append(grosses)
````

# Scraping multiple pages

<h5>
To scrape multiple pages of content, and import the data into a .csv file
</h5>


First define a variable named ```pages``` which are quite explicit about our intentions. Since each page contains 50 movies and there are almost 10,000 movies, we will use arange() to create "pages" variable.
But note that some of the last pages were not scraped due to the url error but we think that it's find and we can work with it.


````python
pages = np.arange(1,9952,50)
````

````python
movies_name = []
year = []
time = []
rating =[]
metascore = []
votes = []
gross = []
````

````python
column_feartures = ['Name of Movie',
                    'Year of Release',
                     'Watch time',
                     'Rating',
                     'Meta Score',
                     'Votes',
                     'Gross Collection($)'
                     ]
````

````python
import csv

csv_file = open('movieUncleaned5.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(column_feartures) # columns of the csv files
````

````python
for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=10000,&sort=user_rating,desc&start="+str(page)+"&ref_=adv_nxt")
    soup = BeautifulSoup(page.text,'lxml')
    movie_data = soup.findAll('div', attrs = {'class': 'lister-item mode-advanced'})
    for store in movie_data:
        name = store.h3.a.text
        movies_name.append(name)
        
        year_of_release = store.h3.find('span',class_='lister-item-year text-muted unbold').text.replace('(', '').replace(')','') if store.h3.find('span',class_='lister-item-year text-muted unbold') else "****"
        year.append(year_of_release)
        
        runtime = store.p.find('span',class_='runtime').text.replace(' min', '') if store.p.find('span',class_='runtime') else "****"
        time.append(runtime)
        
        rate = store.find('div', class_= "inline-block ratings-imdb-rating").text.replace('\n','') if store.find('div', class_= "inline-block ratings-imdb-rating") else "****"
        rating.append(rate)
        
        score = store.find('span',class_='metascore').text.replace('min','') if store.find('span',class_='metascore') else "****"
        metascore.append(score)
        
        value = store.find_all('span',attrs = {'name': "nv" })
        
        vote = value[0].text if len(value)>1 else '****'
        votes.append(vote)
        
        grosses = value[1].text.replace('M', '').replace('$','') if len(value)>1 else '****'
        gross.append(grosses)


        csv_writer.writerow([name, year_of_release, runtime, rate, score, vote, grosses])
````
