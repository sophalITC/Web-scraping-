```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
```

### Web Scrapping from  Ary Phone Store

```python
# create empty array list
phone_samsung = []
phone_price = []
```

```python
# get url from website
page = requests.get("https://arystorephone.com/cate/samsung/").text
soup = bs(page,'lxml')
#soup.prettify()
data = soup.find(class_='products elementor-grid columns-6')
phone_data = data.find_all('li')
phone_data
```

```python
for store in phone_data:
    # find model of phone
    name = store.find('h2',class_='woocommerce-loop-product__title').text
    phone_samsung.append(name)
    # find detail information
    price = store.bdi.text
    #price = store.find('span', class_='woocommerce-Price-amount amount').text
    phone_price.append(price)
```

```python
# print phone_price in a list
phone_price
```

```python
print('There are {} data of price in this page'.format(len(phone_price)))
```

```python
# print phone_samsung in a list
phone_samsung
```

```python
print('There are {} model in this page'.format(len(phone_samsung)))
```

### web scrapping from Ary's phone store

```python
# in this line, we use requests to request the website Ary Store Phone
r = requests.get("https://arystorephone.com/cate/apple/?swoof=1&product_cat=iphone&really_curr_tax=43-product_cat")
# in this line, we use beautifulSoup 
soup = bs(r.content)
# in this line, we use find_all to find the data in class "woocommerce-LoopProduct-link woocommerce-loop-product__link"
iphone = soup.find_all('a', class_="woocommerce-LoopProduct-link woocommerce-loop-product__link")
# in this line, we make list to store the data 
phones = []
# in this line, we make variable store to loop through elements in class "woocommerce-LoopProduct-link woocommerce-loop-product__link"
for store in iphone:
    # in this line, we make variable iphone's name to asign element in h2 in website
    name = store.h2.text
    # for this line, we get iphone's price at the bdi location in website
    price = store.bdi.text
    # for this line, the output have a lots of useless word like these words below so i use loop to delete these words 
    if "Apple MagSafe Battery Pack" in name:
        break
    else:
        print(name)
        print(price)
        # in this line, we use pandas to store data in dataFrame
        phones.append([name, price])
        # in this line, we store data to csv file name iphoneModelPrice1
df1 = pd.DataFrame(phones, columns=['Model', 'Price']) 
df1.to_csv('iphoneModelPrice1.csv', index=False )
```
