## 1. Import Libraries

```python
import pandas as pd
import numpy as np
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as soup
```

## 2. HTTP Request

```python
url = "https://www.khmer24.com/en/c-fridges-and-freezers.html?&per_page="
```

```python
req = Request(url , headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"})
webpage = urlopen(req).read()
page = soup(webpage, "html.parser")
```

## 3. Getting Informations of the first page

### Find all item's detail in first page.

```python
all_item = []
item_title = []
item_price = []
description = []
shop_location = []
post_duration = []

for x in page.find_all('div', class_ = 'item-detail'):
    all_item.append(x)
all_item
```

### Find all item's title in first page.

```python
   
for x in page.find_all('h2',{'class':'item-title'}):
    item_title.append(x.text)
item_title
```

### Find all item's price in first page.

```python
 
for x in page.find_all('span',{'class':'price'}):
    item_price.append(x.text)
item_price
```

### Find all item's description in first page.

```python
 
for x in page.find_all('p',{'class':'description'}):
    description.append(x.text)
description
```

### Find all shop's location in first page.  

```python
 
for x in page.find_all('ul', {'class':'list-unstyled summary'}):
    shop_location.append(x.li.text)
shop_location
```

### Find all post duration in first page.    

```python
 
for x in page.find_all('ul', {'class':'list-unstyled summary'}):
    post_duration.append(x.time.text)
post_duration
```

  
## 4. First look of some informations that we scrapped from first page 

```python

item_title = []
item_price = []
description = []
shop_location = []
post_duration = []

for x in page.find_all('h2',{'class':'item-title'}):
    item_title.append(x.text)
for x in page.find_all('span',{'class':'price'}):
    item_price.append(x.text)
for x in page.find_all('p',{'class':'description'}):
    description.append(x.text) 
for x in page.find_all('ul', {'class':'list-unstyled summary'}):
    shop_location.append(x.li.text)
for x in page.find_all('ul', {'class':'list-unstyled summary'}):
    post_duration.append(x.time.text)
 
```

```python
data = {'Title':item_title, 'Price':item_price, 'Description':description, 'Shop Location':shop_location, 'Duration':post_duration}
items_data = pd.DataFrame(data)
items_data
```

## 5. Scrapping all informations from all available pages.

```python
pages = np.arange(0, 1700, 50) #In Khmer24 it has 50 item's details in one page.

item_title = []
item_price = []
description = []
shop_location = []
post_duration = []



for page in pages:
    url = "https://www.khmer24.com/en/c-fridges-and-freezers.html?&per_page="+str(page)
    req = Request(url , headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"})
    webpage = urlopen(req).read()
    page = soup(webpage, "html.parser")
    print(url, 'SUCESSFUL')
    
    for x in page.find_all('h2',{'class':'item-title'}):
        item_title.append(x.text)
    for x in page.find_all('span',{'class':'price'}):
        item_price.append(x.text)
    for x in page.find_all('p',{'class':'description'}):
        description.append(x.text) 
    for x in page.find_all('ul', {'class':'list-unstyled summary'}):
        shop_location.append(x.li.text)
    for x in page.find_all('ul', {'class':'list-unstyled summary'}):
        post_duration.append(x.time.text)
data = {'Title':item_title, 'Price':item_price, 'Description':description, 'Shop Location':shop_location, 'Duration':post_duration}
items_data = pd.DataFrame(data)

items_data
```

## FINAL. Load all data that we scrapped to the CSV file.
